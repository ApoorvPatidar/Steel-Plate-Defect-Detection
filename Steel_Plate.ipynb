Feature Droping and Target

```python
feature = train_df.drop(["id", "Pastry", "Z_Scratch", "K_Scatch", "Stains", "Dirtiness", "Bumps", "Other_Faults"], axis=1)
# axis 1 refers to the columns

print(feature.shape)
feature.head()
```

```python
test_df.head().T
# t is used for viewing the datafram in transpose format
```

```python
test = test_df.drop(['id'], axis=1)
test.shape
```

```python
# new dataframe is made with the values that will be predicted

target = train_df[["Pastry", "Z_Scratch", "K_Scatch", "Stains", "Dirtiness", "Bumps", "Other_Faults"]]  # used to make a list of columns
# double brackets used to get a subset of the dataframe

print(target.shape)
target.head()
```

Modelling

```python
# This diversity library usage can lead to improved generalization and robustness, especially when combining predictions from multiple models.

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.model_selection import cross_val_score

xgb_models = []
lgbm_models = []
catboost_models = []

# xgb_model is an instance of XGBClassifiers


for i in target.columns:
    xgb_model = XGBClassifier(random_state = 2400, n_estimators = 200, learning_rate = 0.03)
    xgb_model.fit(feature, target[i])  # its trained to understandt the relation between target[i](dependent) and the features(independent) here features after droping the following features 
    xgb_models.append(xgb_model)
    print(f"\nXGB Model for {i}, Area under the ROC Curve = ", f"{cross_val_score(xgb_model, feature, target[i], cv=5, scoring='roc_auc').mean():.5f}")
    
    lgbm_model = LGBMClassifier(random_state=2403, verbose=-1, n_estimators = 200, learning_rate = 0.03)
    lgbm_model.fit(feature, target[i])
    lgbm_models.append(lgbm_model)
    print(f"LGBM Model for {i}, Area under the ROC Curve = ", f"{cross_val_score(lgbm_model, feature, target[i], cv=5, scoring='roc_auc').mean():.5f}")

    catboost_model = CatBoostClassifier(random_state=2403, verbose=0, iterations = 100, learning_rate = 0.1)
    catboost_model.fit(feature, target[i])
    catboost_models.append(catboost_model)
    print(f"CatBoost Model for {i}, Area under the ROC Curve = ", f"{cross_val_score(catboost_model, feature, target[i], cv=5, scoring='roc_auc').mean():.5f}")
```

```python
print("XGB Model, Area under the ROC Curve = ", (0.86429 + 0.95728 + 0.98493 + 0.99185 + 0.88516 + 0.80769 + 0.70293)/7)

print("LGBM Model, Area under the ROC Curve = ", (0.86439 + 0.95817 + 0.98462 + 0.99231 + 0.88423 + 0.80714 + 0.70203)/7)

print("CatBoost Model, Area under the ROC Curve = ", (0.86970 + 0.96026 + 0.98455 + 0.99287 + 0.88484 + 0.80890 + 0.70479)/7)
```

Tuning XGBClassifier for each Target

```python
# Pastry - HyperParameter Tuning

from sklearn.model_selection import GridSearchCV

# these are the set of hyperparameters that are used to tune the model

param_grid = {"n_estimators": [200],      # these are the number of boosting rounds
              "learning_rate": [0.03],    # a low learning rate is generally used as it makes the model more robust and with low learning rate more boosting rounds needed(n_estimators)
              "max_depth": [4],           # A higher max_depth allows the model to capture more complex patterns in the data but increases the risk of overfitting.
              "min_split_loss":[1],             
              "colsample_bynode": [0.6]   # fraction of features to be used in each boosting round i.e is 60% here
              } 


model_xgb = XGBClassifier(booster = 'gbtree')

grid_search = GridSearchCV(model_xgb, param_grid, cv=5, scoring='accuracy')
#(base estimator, hyperparameters, cross validation, scoring metric)
# cv(cross validation) is the number of partitions used for training and validation

grid_search.fit(feature, target["Pastry"])

print("The best hyperparameters are for Pastry", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)

# best_params_ and best_score_ are the attributes of the grid_search object
```
